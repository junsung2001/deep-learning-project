```python
from PIL import Image
import matplotlib.pyplot as plt

# 이미지 불러오기 (파일 경로를 자신의 이미지 경로로 바꿔주세요)
image = Image.open('train/1/00001_00003_00027.png')

# 이미지 보기
plt.imshow(image)
plt.axis('off')  # 축 제거
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 이미지 불러오기 (OpenCV는 기본적으로 BGR로 읽음)
image = cv2.imread('train/0/00000_00000_00028.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # RGB로 변환

# 이미지 보기
plt.imshow(image)
plt.axis('off')
plt.show()
```
